```python
# @title üé¨ Montaje TikTok con Efecto "P√©rdida de Se√±al" (Glitch)

# 1. Instalar MoviePy
!pip install moviepy

from moviepy.editor import *
import numpy as np
import os

# --- CONFIGURACI√ìN ---
video_files = ["Clip1.mp4", "Clip2.mp4", "Clip3.mp4"]
final_image_file = "Final_image.png"
output_filename = "TikTok_Data_Final.mp4"

# --- FUNCI√ìN: GENERADOR DE RUIDO (GLITCH) ---
def create_static_noise(duration=0.4, fps=24, resolution=(1080, 1920)):
    """
    Genera un clip de video de 'nieve' (ruido aleatorio) para transiciones.
    Resoluci√≥n por defecto: Vertical (1080x1920).
    """
    w, h = resolution

    def make_frame(t):
        # Genera una matriz de p√≠xeles aleatorios (ruido visual)
        # Escala de grises (ruido blanco y negro) o color
        return np.random.randint(0, 256, (h, w, 3), dtype=np.uint8)

    return VideoClip(make_frame, duration=duration).set_fps(fps)

# --- FUNCI√ìN: ZOOM SUAVE (Para la imagen final) ---
def zoom_in_effect(clip, zoom_ratio=0.04):
    def effect(get_frame, t):
        img = clip.get_frame(t)
        h, w = img.shape[:2]
        scale = 1 + (zoom_ratio * t)
        new_w, new_h = int(w * scale), int(h * scale)
        import cv2
        img_resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
        x_c, y_c = new_w // 2, new_h // 2
        return img_resized[y_c - h//2 : y_c + h//2, x_c - w//2 : x_c + w//2]
    return clip.fl(effect)

# --- MONTAJE ---
print("üöÄ Iniciando el ensamblaje de la transmisi√≥n...")

final_clips_sequence = []

# 1. Procesar Videos
for i, filename in enumerate(video_files):
    if os.path.exists(filename):
        print(f"üéûÔ∏è Cargando {filename}...")
        clip = VideoFileClip(filename)

        # Asegurar que sea vertical, si no, redimensionar (opcional, por seguridad)
        # clip = clip.resize(height=1920)

        final_clips_sequence.append(clip)

        # A√±adir efecto de INTERFERENCIA entre clips (pero no despu√©s del √∫ltimo)
        if i < len(video_files) - 1:
            print("   ‚ö° Insertando p√©rdida de se√±al...")
            # Creamos 0.3 segundos de est√°tica
            noise = create_static_noise(duration=0.3, resolution=clip.size)
            # Opcional: A√±adir un sonido de est√°tica si tuvieras un audio
            final_clips_sequence.append(noise)
    else:
        print(f"‚ö†Ô∏è Alerta: No encuentro {filename}")

# 2. Procesar Imagen Final (Cr√©ditos)
if os.path.exists(final_image_file):
    print("üñºÔ∏è A√±adiendo cr√©ditos finales...")
    credits_clip = ImageClip(final_image_file).set_duration(4).set_fps(24)
    # Aplicamos el zoom suave que nos gusta
    credits_clip = zoom_in_effect(credits_clip)

    # Transici√≥n de interferencia antes de los cr√©ditos tambi√©n
    noise = create_static_noise(duration=0.4, resolution=credits_clip.size)
    final_clips_sequence.append(noise)
    final_clips_sequence.append(credits_clip)

# 3. Renderizar
if len(final_clips_sequence) > 0:
    print("\nüì° Uniendo transmisi√≥n intergal√°ctica...")
    final_video = concatenate_videoclips(final_clips_sequence, method="compose")

    # Exportar
    final_video.write_videofile(output_filename, fps=24, codec='libx264', audio_codec='aac')
    print(f"\n‚úÖ ¬°TRANSMISI√ìN COMPLETADA! Descarga: {output_filename}")
else:
    print("‚ùå Error: No hay material para montar.")
```
